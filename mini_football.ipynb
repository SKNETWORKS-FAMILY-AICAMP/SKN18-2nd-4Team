```python
import pandas as pd
import os

# 데이터프레임들을 저장할 빈 딕셔너리를 만듭니다.
dataframes = {}

print("데이터프레임 변환 시작...")
# file_list에 있는 각 파일 경로(file_path)에 대해 반복 작업을 수행합니다.
for file_path in file_list:
    # 경로에서 순수 파일 이름만 추출합니다 (예: 'players.csv')
    file_name = os.path.basename(file_path)

    # '.csv' 확장자를 제거하여 데이터프레임의 이름으로 사용합니다 (예: 'players')
    df_name = file_name.replace('.csv', '')

    # pd.read_csv() 함수로 파일을 읽어 DataFrame으로 변환합니다.
    df = pd.read_csv(file_path)

    # 생성된 DataFrame을 딕셔너리에 저장합니다.
    dataframes[df_name] = df

    print(f"'{file_name}'  ➡️  dataframes['{df_name}'] 변환 완료")

print("\n--- 변환 완료! ---")


# 변환이 잘 되었는지 확인해봅니다.
print("\n저장된 데이터프레임 목록:")
print(list(dataframes.keys()))
```

```python
import pandas as pd

def get_season(date):
  # 7월 이상이면, 시즌은 현재 연도에 시작
  if date.month >= 7:
    start_year = date.year
    end_year = date.year + 1
  # 7월 미만(1~6월)이면, 시즌은 이전 연도에 시작
  else:
    start_year = date.year - 1
    end_year = date.year

  # '17/18'과 같은 형식으로 반환
  return f"{str(start_year)[-2:]}/{str(end_year)[-2:]}"
```

```python


# (중요) 'date' 열을 날짜 타입으로 먼저 변환해야 합니다.
games_df['date'] = pd.to_datetime(games_df['date'])

# 'date' 열의 각 날짜에 get_season 함수를 적용
games_df['season'] = games_df['date'].apply(get_season)

games_df.head()
#         date season
# 0 2017-08-15  17/18  (8월이므로 17/18 시즌)
# 1 2018-03-20  17/18  (3월이므로 17/18 시즌)
# 2 2018-11-05  18/19  (11월이므로 18/19 시즌)
```

```python
def player_result(row):
    if row["player_club_id"] == row["home_club_id"]:  # 홈팀 소속 선수
        if row["home_club_goals"] > row["away_club_goals"]:
            return "Win"
        elif row["home_club_goals"] < row["away_club_goals"]:
            return "Lose"
        else:
            return "Draw"
    elif row["player_club_id"] == row["away_club_id"]:  # 원정팀 소속 선수
        if row["away_club_goals"] > row["home_club_goals"]:
            return "Win"
        elif row["away_club_goals"] < row["home_club_goals"]:
            return "Lose"
        else:
            return "Draw"
    else:
        return None
```

```python
GB1_final_df8["player_result"] = GB1_final_df8.apply(player_result, axis=1)
```

```python
player_season_stats = GB1_final_df8.groupby(["season_y", "player_name",'player_id']).agg({
    # 수치형 - 합계
    "goals": "sum",
    "assists": "sum",
    "yellow_cards": "sum",
    "red_cards": "sum",

    # 수치형 - 평균
    "minutes_played": "mean",
    "market_value_in_eur": "mean",
    "squad_size": "mean",
    "average_age": "mean",
    "foreigners_percentage": "mean",
    'national_team_players': "mean",

    # 수치형 - 최대
    "highest_market_value_in_eur": "max",

    # object - 선수 정보
    'name_y':'first',
    "country_of_birth": "first",
    "date_of_birth": "first",
    "foot": "first",
    "height_in_cm": "first",
    "agent_name": "first",

    # object - 포지션
    "net_transfer_record": lambda x: x.mode()[0] if not x.mode().empty else None,
    "position": lambda x: x.mode()[0] if not x.mode().empty else None,
    "sub_position": lambda x: x.mode()[0] if not x.mode().empty else None,
    "player_result": lambda x: (x == "Win").sum()
}).reset_index()
```
